```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('mic2.xlsx',
                  engine = 'openpyxl')
df.head()
```

```python
def custom_cls(x):
    if (0<= x <=20):
        return '0 a 20'
    elif(21<= x <= 30):
        return '21 a 30'
    elif(31<= x <= 40):
        return '31 a 40'
    else:
        return 'Mayor a 40'
```

```python
df['Age Range'] = df.apply(lambda tbl: custom_cls(tbl['Age']), axis = 1)
df.head(2)
```

```python
counts = df['Age Range'].value_counts()
counts = counts.to_frame()
counts
```

```python
counts['%'] = counts.iloc[:, :]/counts.loc[:,'Age Range'].sum(axis=0)*100
counts
```
